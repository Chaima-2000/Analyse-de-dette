## Séries Choisies et Leur Pertinence :
### External debt stocks, total (DOD, current US$)

#####  Cette série indique le stock total de la dette extérieure d'un pays, c'est-à-dire la somme des emprunts impayés à un moment donné
#### Principal repayments on external debt, long-term (AMT, current US$)

#####  Cette série représente les paiements de principal sur la dette extérieure à long terme. 
#### Interest payments on external debt, total (INT, current US$)

#####  Cette série indique les paiements d'intérêts totaux sur la dette extérieure
#### Net flows on external debt, total (NFL, current US$)

#####  Les flux nets de la dette extérieure mesurent les entrées et sorties de capitaux liés à la dette, y compris les nouveaux emprunts et les remboursements

```python

# Séries pertinentes pour l'analyse
relevant_series = [
    "External debt stocks, total (DOD, current US$)",
    "Principal repayments on external debt, long-term (AMT, current US$)",
    "Interest payments on external debt, total (INT, current US$)",
    "Net flows on external debt, total (NFL, current US$)",
]

# Filtrer les données pour les séries pertinentes
filtered_data = data[data['Series Name'].isin(relevant_series)]

# Afficher un aperçu
print(filtered_data['Series Name'].unique())
```

```python
values = filtered_data['value']

# Calculer les statistiques
mean_value = np.mean(values)        # Moyenne
median_value = np.median(values)    # Médiane
std_dev = np.std(values)            # Écart type
min_value = np.min(values)          # Valeur minimale
max_value = np.max(values)          # Valeur maximale

# Afficher les résultats
print(f"Moyenne : {mean_value}")
print(f"Médiane : {median_value}")
print(f"Écart type : {std_dev}")
print(f"Valeur minimale : {min_value}")
print(f"Valeur maximale : {max_value}")
```

## Division de l'ensemble de données en un ensemble d'entrainement et un ensemble de test 


```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Division du dataset en variables d'entrée (X) et variable cible (y)

debt_stock = filtered_data[filtered_data['Series Name'] == 'External debt stocks, total (DOD, current US$)']
# Préparer les données pour le modèle (X = années, y = valeurs)
X = debt_stock[['Year']]  # Variable indépendante (Année)
y = debt_stock['value']   # Variable dépendante (Stock de la dette
debt_stock.head()
```

```python
# Diviser les données en un ensemble d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print("Taille de l'ensemble d'entraînement :", X_train.shape)
print("Taille de l'ensemble de test :", X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
# Créer et entraîner le modèle de régression linéaire
model = LinearRegression()
model.fit(X_train, y_train)
```

```python

# Faire des prédictions sur les données de test
y_pred = model.predict(X_test)
print(y_pred)
```

## Evaluation 

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Calculer le MSE
mse = mean_squared_error(y_test, y_pred)
print(f"MSE (Erreur quadratique moyenne) : {mse}")
```

```python
# Calculer R²
r2 = model.score(X_test, y_test)
print(f"R² (Coefficient de détermination) : {r2}")
```

```python
from sklearn.metrics import mean_absolute_error

# Calculer le MAE
mae = mean_absolute_error(y_test, y_pred)
print(f"MAE (Erreur absolue moyenne) : {mae}")
```
